```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from pandas.plotting import parallel_coordinates

plt.style.use("seaborn-colorblind")

%matplotlib inline
```

```python
df = pd.read_csv("data.csv")
df = df[list(df.columns)[1:]]
print(df.shape)
```

## Поиск недостающих данных

```python
na = df.isnull().sum()
na[na > 0]
```

```python
# Индексы недостающих данных
np.nonzero(pd.isnull(df.values).any(axis=1))
```

```python
# Выбрасываем недостающие данные
df = df.dropna(axis=0)
print(df.shape)
df.head()
```

## Подготовка данных

```python
X = df.iloc[:, :-1]
y = df[["cd/A"]]
print(X.shape, y.shape)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score, train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=7
)
```

# Модели машинного обучения
## Случайный лес

```python
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor(n_estimators=12, random_state=5)
forest.fit(X_train, y_train.values.ravel())
y_pred = forest.predict(X_test)

print("Коэффициент детерминации: ", forest.score(X_test, y_test))
print("Среднеквадратичная ошибка модели: ", np.sqrt(mean_squared_error(y_test, y_pred)))
```

## XGB

```python
from xgboost import XGBRegressor

xgb = XGBRegressor(random_state=4)
xgb.fit(X_train, y_train.values.ravel())
y_pred = xgb.predict(X_test)

print("R-squared test: ", xgb.score(X_test, y_test))
print("RMSE: ", np.sqrt(mean_squared_error(y_test, y_pred)))
```
